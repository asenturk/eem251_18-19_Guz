Regular Expressions Düzenli (kurallı) İfadeler

\d bir rakam demek  
\w harf rakam _ yi içermektedir.


```python
#dosyadaki rakamlar
import re #regular expressions
regex=re.compile(r"\d")
bulunanlar=regex.findall(makale)
#print(bulunanlar)
print(" ". join(bulunanlar))
```

2 rakamlı sayıları bulalım

```python
import re #regular expressions
regex=re.compile(r"\d\d")
bulunanlar=regex.findall(makale)
print(" ". join(bulunanlar))

print("4 rakamlı sayılar")
regex=re.compile(r"\d\d\d\d")
bulunanlar=regex.findall(makale)
print(" ". join(bulunanlar))

#en az 2 rakamlı en çok 4 rakamlı sayılar
print("2,3,4 rakamlı sayılar")
regex=re.compile(r"\d{2,4}")
bulunanlar=regex.findall(makale)
print(" - ". join(bulunanlar))
```

```python
import re
#uzunluğu en az 5 en fazla 7  karakterden oluşmuş ifadeler
regex=re.compile(r"\w{5,7}")
bulunanlar=regex.findall(makale)
print(" ". join(bulunanlar))
```

```python
import re
# whitespace: boşluk yeni satır tab gibi ifadeler
regex=re.compile(r"\s")
bulunanlar=regex.findall(makale)
print(bulunanlar)
```

\b: kelime sınırı

```python
import re
regex=re.compile(r"\b\w\w?\b")
bulunanlar=regex.findall(makale)
print(" ". join(bulunanlar))
```

```python
import re
#a ile başlayan en fazla 3 harfli ifadeler
regex=re.compile(r"\ba\w?\w?\b")
bulunanlar=regex.findall(makale)
print(" ". join(bulunanlar))
```

```python
import re
#a ile başlayan  3 harfli kelimeler
regex=re.compile(r"\ba\w{2}\b")
bulunanlar=regex.findall(makale)
print(" ". join(bulunanlar))
```

```python
import re
#3 harfli kelimeler, rakam yok. Türkçe karakter destekli
regex=re.compile(r"\b[a-züğışçöA-ZÜĞİŞÇÖ]{3}\b")
bulunanlar=regex.findall(makale)
print(bulunanlar)
```

```python
#\b kelime sınırı
#[a-dç] a,b,c,ç,d harflerindeni 1 tanesi
#\w{4} 4 tane karakter veya rakam veya _ karakteri
regex=re.compile(r"\b[a-dç]\w{4}\b")
bulunanlar=regex.findall(makale)
print(bulunanlar)
```

a, b, c, d ile başlaMAyan 4 harfli kelimeler

```python
# köşeli parantezin içine yazılan ^, değil anlamına gelir.
# abcd harfleri ve rakamla başlamayan 4 harfli kelimeler
import re
regex=re.compile(r"\b[^abcd 0-9]\w{3}\b")
bulunanlar=regex.findall(makale)
print(bulunanlar[:100])
```

python ile başlayan pdf dosyaları  
^: başlama ifadesi  
$: bitiş ifadesi  

```python
import os, re
regex=re.compile(r"^[pP]ython.+?.pdf$")
print("Dosyalar")
files=os.listdir("./dosyalar/")
print(files)
for file in files:
    bulunan_dosya=regex.findall(file)
    print(bulunan_dosya)
```

| veya operatörü

```python
import re
regex=re.compile(r"(I love )(dog|cat)")
bulunanlar=regex.findall("I love dog, I love fish, I love bird, I love cat")
print(bulunanlar)
```

gruplama

```python
email_address = 'Please contact us at: support@datacamp.com'
match = re.search(r'([\w\.-]+)@([\w\.-]+)', email_address)
if match:
    print(match.group()) # The whole matched text
    print(match.group(1)) # The username (group 1)
    print(match.group(2)) # The host (group 2)
```
