```python
# import modules
import RPi.GPIO as GPIO
import time

# setup pins
GPIO.setmode(GPIO.BOARD)
GPIO.setup(3, GPIO.OUT)
GPIO.setup(5, GPIO.IN)

# loop 5 times
for i in range(5):
    
    # flash output pin 3
    GPIO.output(3, GPIO.HIGH)
    time.sleep(1)
    GPIO.output(3, GPIO.LOW)
    time.sleep(1)
    
    # read input pin 5
    if GPIO.input(5) == GPIO.HIGH:
        print("Pin 5 is on")
    else:
        print("Pin 5 is off") 
```

```python
# import modules
import RPi.GPIO as GPIO
import time

# setup pins
GPIO.setmode(GPIO.BOARD)
GPIO.setup(3, GPIO.OUT)
GPIO.setup(5, GPIO.OUT)
GPIO.setup(8, GPIO.IN)

while(1):
    if GPIO.input(8) == GPIO.HIGH:
        GPIO.output(3, GPIO.HIGH)
        GPIO.output(5, GPIO.LOW)
    else:
        GPIO.output(5, GPIO.HIGH)
        GPIO.output(3, GPIO.LOW)
    time.sleep(0.5)
```

```python
# import modules
import RPi.GPIO as GPIO
import time

# setup pins
GPIO.setmode(GPIO.BOARD)
GPIO.setup(8, GPIO.IN)
pinler=[3,5,7,11,13,15]
for i in pinler:
    GPIO.setup(i, GPIO.OUT)

while(1):
    if GPIO.input(8) == GPIO.HIGH:
        for i in pinler:
            GPIO.output(i, GPIO.LOW)
        for i in pinler:
            GPIO.output(i, GPIO.HIGH)
            time.sleep(1)
            GPIO.output(i, GPIO.LOW)
            if GPIO.input(8) == GPIO.LOW:
            break
    else:
        for i in pinler:
            GPIO.output(i, GPIO.LOW)
        for i in pinler[::-1]:
            GPIO.output(i, GPIO.HIGH)
            time.sleep(1)
            GPIO.output(i, GPIO.LOW)
            if GPIO.input(8) == GPIO.HIGH:
                break
```

```python
# import modules
import RPi.GPIO as GPIO
import time

# setup pins
GPIO.setmode(GPIO.BOARD)
#3. 5. ve 7. pinler output olarak ayarlandi.
GPIO.setup(3, GPIO.OUT) 
GPIO.setup(5, GPIO.OUT)
GPIO.setup(7, GPIO.OUT)


a=[3,5,7]

for i in a:
    #i. outputu lojik 1 yap
    GPIO.output(i, GPIO.HIGH)
    # 1 sn bekleme süresi
    time.sleep(1)

for i in a[::-1]:
    GPIO.output(i, GPIO.LOW)
    time.sleep(1)
```

```python
# import modules
import RPi.GPIO as GPIO
import time

# setup pins
pinler=[3,5,7,11,13,15,19,21]

GPIO.setmode(GPIO.BOARD)
for pin in pinler: 
    GPIO.setup(pin, GPIO.OUT)


while(True):
    for i in range(0,len(pinler),2):
        GPIO.output(pinler[i], GPIO.HIGH)
        GPIO.output(pinler[i+1], GPIO.HIGH)
        time.sleep(1)
    
    for i in range(0,len(pinler),2):
        GPIO.output(pinler[i], GPIO.LOW)
        GPIO.output(pinler[i+1], GPIO.LOW)
        time.sleep(1)
```
